# Directly training the ML Models without any Dimentionality reduction or Feature transformations

```python
import pandas as pd

# Load the training dataset
training_set = pd.read_csv('training_set.csv')
```

```python
training_set.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
training_set
```

```python
from sklearn.model_selection import train_test_split

X = training_set.drop('Y', axis=1)
y = training_set['Y']

X_train, X_validation, y_train, y_validation = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import numpy as np

# Initialize the Random Forest classifier
# Note: Hyperparameters are set to default values and should be tuned
random_forest = RandomForestClassifier(random_state=0)

# Train the Random Forest classifier
random_forest.fit(X_train, y_train)

# Predict on the validation set
y_pred = random_forest.predict(X_validation)
y_prob = random_forest.predict_proba(X_validation)[:, 1]

# Evaluate the model's performance
print("Random Forest Performance:")
print("Classification Report:\n", classification_report(y_validation, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_validation, y_pred))
print("ROC AUC Score:", roc_auc_score(y_validation, y_prob))

# Feature importances
importances = random_forest.feature_importances_
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")
for f in range(X_train.shape[1]):
    print(f"{f + 1}. feature {indices[f]} ({importances[indices[f]]})")
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
import lightgbm as lgb
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score


# Define the models to train
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Support Vector Machine": SVC(probability=True),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier(),
    "Naive Bayes": GaussianNB(),
    "k-NN": KNeighborsClassifier(),
    "LightGBM": lgb.LGBMClassifier(verbose=-1),
    "Decision Tree": DecisionTreeClassifier()
}

# Dictionary to store the results
results = {}

# Train and evaluate each model
for name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)

    # Make predictions on the validation set
    predictions = model.predict(X_validation)

    # For models that can estimate probabilities
    if name in ["Support Vector Machine", "Random Forest", "Gradient Boosting", "LightGBM"]:
        probabilities = model.predict_proba(X_validation)[:,1]
        roc_auc = roc_auc_score(y_validation, probabilities)
    else:
        roc_auc = "N/A"  # For models that do not support probability predictions

    # Calculate metrics
    accuracy = accuracy_score(y_validation, predictions)
    precision = precision_score(y_validation, predictions)
    recall = recall_score(y_validation, predictions)
    f1 = f1_score(y_validation, predictions)

    # Store the results
    results[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'ROC AUC': roc_auc
    }

# Display the results
for model in results:
    print(f"Model: {model}")
    for metric, value in results[model].items():
        print(f"{metric}: {value}")
    print("\n")
```

PCA and feature scaling can lead to varying effects on model performance, depending on the nature of the algorithm and the dataset. While tree-based models like Random Forest and Gradient Boosting are generally robust to feature scale, they can still benefit from the variance-focused feature selection of PCA. Models like SVM and k-NN, which rely on distance calculations, can show significant performance changes when feature scaling is applied. However, care must be taken with dimensionality reduction techniques like PCA, as they can sometimes discard information that may be valuable for prediction, especially for models that rely on the raw features, like Naive Bayes and Decision Trees.
